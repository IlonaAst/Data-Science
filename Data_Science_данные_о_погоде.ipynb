```python
# Подключаем библиотеку с математическими функциями
import numpy as np
```

```python
# Необходимо подогнать косинусоиду по ширине => период от 1 до 366 к периоду от 0 до 2pi
# День в году в радианах = (dayofyear -1)/366*2pi
# Косинус от дня в году = cos(день в году в радианах из диапазона от 0 до 2pi)
data['cos_dayofyear'] = np.cos((data['dayofyear'] - 1) / 366 * 2 * np.pi)
```

# Разделение данных на обучающую и тестовую выборку
data_train и data_test

```python
data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']
```

```python
plt.plot(data_train['date'], data_train['T']) # синяя часть данных - обучение
plt.plot(data_test['date'], data_test['T'])  # оранжеыая часть данных - тестирование
```

```python
# Чтобы обучить модель, нужно ещё раз разделить выборку на фичи (признаки, features, X) и таргет (целевую переменную y)
# 2 датасета train-test x 2 ауфегкуы-ефкпуе - создаем 4 переменные X_train, y_train, X_test, y_test 
# Будем делать прогноз только на одном факторе - номере дня в году ['dayofyear']
# Но модель ожидает, что ей на вход придет двумерная таблица с многими факторами,
# Поэтому создаем полноценную таблицу pandas DataFrame из колонки T

X_train = pd.DataFrame()
X_train['cos_dayofyear'] = data_train['cos_dayofyear']

X_test = pd.DataFrame()
X_test['cos_dayofyear'] = data_test['cos_dayofyear']

# y оставляем столбцом как есть
y_train = data_train['T']
y_test = data_test['T']
```

# 1-я модель: Линейная регрессия (Linear Regression)

z = acos(dayofyear) + b

С ее помощью будем подгонять коэффициент при косинусе так, чтобы растянуть его в высоту до уровня наших данных

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# Обучаем модель: подгоняем модель по данным X_train и говорим правильные ответы y_train
model.fit(X_train, y_train)
```

```python
# Проверяем, какую мат. модель построила регрессия по данным
pred_train = model.predict(X_train)
# Распечатаем прогноз на данных, которые модель еще не видела
pred_test = model.predict(X_test)

# Строим графики
plt.plot(data_train['date'], y_train) 
plt.plot(data_test['date'], y_test) 
plt.plot(data_train['date'], pred_train, color = 'yellow') 
plt.plot(data_test['date'], pred_test) 
```

```python
# Проверяем качество численно. mean_absolute_error - средняя сумма отклонений (меньше => лучше)

from sklearn.metrics import mean_absolute_error

print('Средняя ошибка на тренировочной выборке =', mean_absolute_error(y_train, pred_train))
print('Средняя ошибка на тестовой выборке =', mean_absolute_error(y_test, pred_test))
```

```python
# Коэффициенты a и b, которые подобрала модель
model.coef_, model.intercept_
```

# 2-я модель: Дерево решений (Decision Tree)

Здесь не нужно косинусное преобразование. Будем делать прогноз на исходном факторе "день в году"

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Заново разбиваем датасет на Train и Test, чтобы изменения применились
data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

# Из train-test формируем X_train, X_test
X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['dayofyear']

X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear']

y_train = data_train['T']
y_test = data_test['T']

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

# Делаем прогноз
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# Формируем графики
plt.figure(figsize=(20, 5))
plt.scatter(data_train['date'], y_train, label = 'Data train') 
plt.scatter(data_test['date'], y_test, label = 'Data test') 
plt.scatter(data_train['date'], pred_train, label = 'Predict train') 
plt.scatter(data_test['date'], pred_test, label = 'Predict test') 
plt.legend()

# Определяем величину ошибки
print('Средняя ошибка на тренировочной выборке =', mean_absolute_error(y_train, pred_train))
print('Средняя ошибка на тестовой выборке =', mean_absolute_error(y_test, pred_test))
```

```python
# Качество на тестовой выборке немного ухудшилось с 4,57 до 4,71 градусов
# Качество на тренировочной выборке немного улучшилось с 4,3 до 3,62 градусов
```

* Модель слишком сильно зазубрила ответы, это называется 
переобучение. Это видно из того, что на тренировочной выборке ошибка меньше, чем на тестовой.
* Исправляем ее - задаем ограничения для модели, чтобы она стала проще и не смогла зубрить.
* В случае дерева мы можем играть с глубиной max_depth или количеством листьев.

# Модель 2.2. Меняем параметры модели "Дерево решений"

```python
# Заново разбиваем датасет на Train и Test, чтобы изменения применились
data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

# Из train-test формируем X_train, X_test
X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['dayofyear']

X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear']

y_train = data_train['T']
y_test = data_test['T']

model = DecisionTreeRegressor(max_depth = 6)
model.fit(X_train, y_train)

# Делаем прогноз
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# Формируем графики
plt.figure(figsize=(20, 5))
plt.scatter(data_train['date'], y_train, label = 'Data train') 
plt.scatter(data_test['date'], y_test, label = 'Data test') 
plt.scatter(data_train['date'], pred_train, label = 'Predict train') 
plt.scatter(data_test['date'], pred_test, label = 'Predict test') 
plt.legend()

# Определяем величину ошибки
print('Средняя ошибка на тренировочной выборке =', mean_absolute_error(y_train, pred_train))
print('Средняя ошибка на тестовой выборке =', mean_absolute_error(y_test, pred_test))
```

```python
# Качество на тестовой выборке немного улучшилось с 4,71 до 4,58 градусов
# Качество на тренировочной выборке немного ухудшилось с 3,62 до 3,77 градусов - модель меньше зубрила
```

# Модель 2.3. Перебираем параметр глубины модели "Дерево решений" для ее улучшения

```python
data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['dayofyear']

X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear']

y_train = data_train['T']
y_test = data_test['T']

# Ищем лучший вариант по параметру глубины max_depth
pred_train2 = pd.DataFrame()
pred_test2 = pd.DataFrame()
min_error = 5
for i in range(2,15):
  model = DecisionTreeRegressor(max_depth = i)
  model.fit(X_train, y_train)

# Делаем прогноз
  pred_train = model.predict(X_train)
  pred_test = model.predict(X_test)
  error = mean_absolute_error(y_test, pred_test)
  if min_error > error:
    pred_train2 = pred_train
    pred_test2 = pred_test
    min_error = error

# Определяем величину ошибки
print('Средняя ошибка на тренировочной выборке =', mean_absolute_error(y_train, pred_train2))
print('Средняя ошибка на тестовой выборке =', mean_absolute_error(y_test, pred_test2))
```

```python
# Качество на тестовой выборке не изменилось = 3,77 градусов
# Качество на тренировочной выборке не изменилось = 4,58 градусов
```

# Модель 1.2. Включаем в модель "Линейная регрессия" два признака:
* cos_dayofyear
* sin_dayofyear

```python
# Новые признаки
data['cos_dayofyear'] = np.cos((data['dayofyear'] - 1) / 366 * 2 * np.pi)
data['sin_dayofyear'] = np.sin((data['dayofyear'] - 1) / 366 * 2 * np.pi)

data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

# Из train-test формируем X_train, X_test
X_train = pd.DataFrame()
X_train = data_train[['cos_dayofyear', 'sin_dayofyear']]

X_test = pd.DataFrame()
X_test = data_test[['cos_dayofyear', 'sin_dayofyear']]

y_train = data_train['T']
y_test = data_test['T']

model = LinearRegression()
model.fit(X_train, y_train)

# Делаем прогноз
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# Формируем графики
plt.figure(figsize=(20, 5))
plt.scatter(data_train['date'], y_train, label = 'Data train') 
plt.scatter(data_test['date'], y_test, label = 'Data test') 
plt.scatter(data_train['date'], pred_train, label = 'Predict train') 
plt.scatter(data_test['date'], pred_test, label = 'Predict test') 
plt.legend()

# Определяем величину ошибки
print('Средняя ошибка на тренировочной выборке =', mean_absolute_error(y_train, pred_train))
print('Средняя ошибка на тестовой выборке =', mean_absolute_error(y_test, pred_test))
```
